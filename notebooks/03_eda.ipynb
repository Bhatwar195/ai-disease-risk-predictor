### Import Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Load Cleaned Datasets

```python
heart=pd.read_csv("C://Users//JOHN//Desktop//ML_projects//ai-disease-risk-predictor//data//raw_data/heart_clean.csv")
diabetes=pd.read_csv("C://Users//JOHN//Desktop//ML_projects//ai-disease-risk-predictor//data//raw_data/diabetes_clean.csv")
```

## Why Exploratory Data Analysis?

EDA helps in:
- Understanding feature distributions
- Identifying relationships with the target variable
- Supporting feature selection decisions
- Providing visual insights for presentation

### Heart Disease: Age Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(heart["age"], kde=True)
plt.title("Age Distribution - Heart Disease Dataset")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### Heart Disease: Cholesterol vs Target

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=heart["target"], y=heart["chol"])
plt.title("Cholesterol Levels vs Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol")
plt.show()
```

### Higher cholesterol is associated with higher heart disease risk.

### Diabetes: Glucose vs Outcome

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=diabetes["target"], y=diabetes["glucose"])
plt.title("Glucose Levels vs Diabetes Risk")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Glucose")
plt.show()
```

### Glucose levels are significantly higher in diabetic patients.

### Diabetes: BMI Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(diabetes["bmi"], kde=True)
plt.title("BMI Distribution - Diabetes Dataset")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

### Higher BMI values indicate obesity, a known diabetes risk factor.

### Correlation Heatmap (Heart Dataset)

```python
plt.figure(figsize=(8,6))
sns.heatmap(heart.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Heart Disease Dataset")
plt.show()
```

## Key Insights from EDA

### Heart Disease Dataset
- Age, cholesterol, and blood pressure show correlation with heart disease
- Heart disease cases increase in middle-aged and elderly groups

### Diabetes Dataset
- Glucose level is the strongest indicator of diabetes
- Higher BMI increases diabetes risk
- Age also contributes to disease likelihood
